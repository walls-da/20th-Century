### Text Mining 20th Century

```python
# Import libraries

from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import nltk
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

```python
#Import 20th Century txt file

with open('20th_century.txt', 'r', errors='ignore') as file: 
   data = file.read().replace( '\n', ' ')
```

```python
# Tokenize sentences

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent)
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
#Check frequency of words

from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words)
```

```python
#Finding the top ten most common words

dist_words.most_common(10)
```

```python
# Frequency Distribution Bar Chart

top10 = dist_words.most_common(10)
words, counts = zip(*top10)

plt.figure(figsize=(10, 4))
plt.bar(words, counts, color='skyblue')
plt.title('Top 10 Most Common Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Removing stopwords in words

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

filtered_words = []
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
# Create new FreqDist for filtered–words

dist_words_filter = FreqDist(filtered_words)
print (dist_words_filter)
```

```python
# Updated Frequency Distribution Bar Chart

filtered_10 = dist_words_filter.most_common(10)
words, counts = zip(*filtered_10)

plt.figure(figsize=(10, 4))
plt.bar(words, counts, color='teal')
plt.title('Top 10 Most Common Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Removing punctuation

# Substitute all punctuation marks with a space

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                        " ",        # Replace all non-letters with spaces
                        str(filtered_words))
```

```python
# Retokenizing words

tokenized_word_2 = word_tokenize(sans_punc)
print (tokenized_word_2)
```

```python
# Create new frequency list

dist_words_filter_2 = FreqDist(tokenized_word_2)
```

```python
dist_words_filter_2
```

```python
# Updated Frequency Distribution Bar Chart

dist_words_filtered_10 = dist_words_filter_2.most_common(10)
words, counts = zip(*dist_words_filtered_10)

plt.figure(figsize=(10, 4))
plt.bar(words, counts, color='green')
plt.title('Top 10 Most Common Words')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Bar chart comparison 

3 bar charts were made: 1 with stopwords and two without. 
The first and second charts were pretty similiar because they both contained punctuation rather than words. The first chart without stopwords actually contained more punctuation than the original bar chart. 

The last bar chart consists of only words or letters. I think having "the" still included in the freqdist doesn't make sense because it's a very common word and not very useful for evaluation. A single letter, "s" is also not very useful -- it could be from a possessive or contraction.

```python
#Tagging POS

# Removing extra common words/letters

new_stopwords = ['And', 'Then', 'n', 't', 's', 'The']

filtered = []
for word in tokenized_word_2:
     if word not in new_stopwords:
        filtered.append(word)
```

```python
%%time
text = TextBlob(str(filtered))
```

```python
# Create a tag list

tags_list = text.tags
```

```python
tags_list
```

```python
# Create dataframe

df_text = pd.DataFrame(tags_list)
df_text.columns = [ 'Words', 'Word type']
df_t=df_text.groupby('Word type').count().reset_index()
top10=df_t.nlargest(10, 'Words')
```

```python
# Create box plot

plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 20):
      sns.barplot(x = 'Words', y = 'Word type',
     saturation = 0.9, data = top20).set_title("20th Century - top 10 word types used")
```

```python
# POS box plot: Top 15 nouns

df_nouns = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNS")]
df_nouns.columns = ['Words', 'Frequency']
x = df_nouns.groupby('Words').count().reset_index()
y = x.sort_values(by = ['Frequency'], ascending=False)
top15 = y.nlargest(15, 'Frequency')
```

```python
# Create noun bar plot

plt.figure(figsize=(20, 10))
with sns.dark_palette("xkcd:blue", 15):
      sns.barplot(x='Words', y='Frequency',
      saturation=0.9, data = top15).set_title("20th Century- Most Frequently Used Nouns")
```

```python
# POS box plot: Top 15 adjective

df_adj = df_text[(df_text['Word type'] == "JJ") | (df_text['Word type'] == "JJR") | (df_text['Word type'] == "JJS")]
df_adj.columns = ['Words', 'Frequency']
x = df_adj.groupby('Words').count().reset_index()
y = x.sort_values(by = ['Frequency'], ascending=False)
top15 = y.nlargest(15, 'Frequency')
```

```python
# Create adjective bar plot

plt.figure(figsize=(20, 10))
with sns.dark_palette("#d0fefe", 15):
      sns.barplot(x='Words', y='Frequency',
      saturation=0.9, data = top15).set_title("20th Century- Most Frequently Used Adjectives")
```

```python
# POS box plot: Top 15 verbs

df_vb = df_text[(df_text['Word type'] == "VB") | (df_text['Word type'] == "VBD") | (df_text['Word type'] == "VBG") | (df_text['Word type'] == "VBN") | (df_text['Word type'] == "VBP") | (df_text['Word type'] == "VBZ") ]
df_vb.columns = ['Words', 'Frequency']
x = df_vb.groupby('Words').count().reset_index()
y = x.sort_values(by = ['Frequency'], ascending=False)
top15 = y.nlargest(15, 'Frequency')
```

```python
# Create verb bar plot

plt.figure(figsize=(20, 10))
with sns.dark_palette("#d0fefe", 15):
      sns.barplot(x='Words', y='Frequency',
      saturation=0.9, data = top15).set_title("20th Century- Most Frequently Used Verbs")
```

### Comparison of POS Charts

There are more verbs repeated with the top one having a count of 200. 
All verbs indicate the past.
Nouns indicate more words illustrating the Holocaust, yet the adjectives and verbs depict no direct mention of it.
Out of context, adjectives and verbs could be about almost anything.

```python
# Extract all words from data

listToStr = ' '.join([str(elem) for elem in filtered])

print(listToStr)
```

```python
# Create a dictionary for count of countries

all_counts = Counter(re.sub(r'\W+', ' ', listToStr).split())
```

```python
# Import country df

df_countries = pd.read_csv("countries_list_20th_century_1.5.csv")
```

```python
df_countries
```

```python
# Create country list

country_list = df_countries['country_name'].str.strip().tolist()
```

```python
country_list
```

```python
country_counts = Counter()
for country in country_list:
    pattern = r'\b' + re.escape(country) + r'\b'
    matches = re.findall(pattern, listToStr)
    if matches:
        country_counts[country.title()] = len(matches)
```

```python
country_counts
```

```python
# Convert to df

df_country_counts = pd.DataFrame(country_counts.items(), columns=['Country', 'Count'])
df_country_counts = df_country_counts.sort_values('Count', ascending=False) 
```

```python
# Plot country frequency

country_freq = df_country_counts.head(10)
sns.barplot(data=country_freq, x='Count', y='Country', hue='Country', legend = False)
plt.title('Frequency of Mentioned Countries')
plt.xlabel('Mentions')
plt.ylabel('Country')
plt.show()
```

### Country Bar Chart analysis

German being the top mentioned country makes sense considering it and its events were most present in the noun top 10 chart
Japan and the US also appeared a lot in the POS charts, so it also makes sense they are high in this chart
Russia was also mentioned in a top 10 chart, yet, it's pretty low in the country chart.
